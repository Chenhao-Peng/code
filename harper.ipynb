```python
import numpy as np
import matplotlib.pyplot as plt

N = 10
x = (1/50)*np.array(range(0,50))
l = 1
py = 0
n = np.array(range(0,N))
for phi in x:
    V=2*l*np.diag(np.cos(py - 2*np.pi*phi*n))
    H=V+np.diag(np.ones(N-1),1)+np.diag(np.ones(N-1),-1)
    H[0,N-1]=1
    H[N-1,0]=1
    eigval = np.linalg.eigvalsh(-H)
    phirep = phi * np.ones(10)
    plt.scatter(phirep,eigval,s = 3, c = 'red')
plt.show()
```

```python

N = 10
x = (1/50)*np.array(range(0,50))
l = 1
py = 0
n = np.array(range(0,N))
up = np.diag(np.ones(N-1),1)
up[0,N-1]=1
down = np.diag(np.ones(N-1),-1)
down[N-1,0]=1
iden = np.diag(np.ones(N))
for phi in x:
    V=2*l*np.diag(np.cos(py - 2*np.pi*phi*n))
    H=np.kron(up,iden) + np.kron(down,iden)+np.kron(iden,up) + np.kron(iden,down) + np.kron(V,iden) + np.kron(iden,V)
    eigval = np.linalg.eigvalsh(-H)
    phirep = phi * np.ones(100)
    plt.scatter(phirep,eigval,s = 3, c = 'red')
plt.show()
```
